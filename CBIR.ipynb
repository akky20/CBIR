# Different Models and Image Retrieval 

```python
import os
import numpy as np
import cv2
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC  # Import SVM
from skimage import exposure
from skimage.feature import hog

# Set the seed for reproducibility
np.random.seed(42)

# Define the base directory where the images are stored
base_directory = 'Dataset/OfficeHomeDataset_10072016'
domains = ['Art', 'Clipart', 'Product', 'Real World']

# Initialize lists to store features and labels
features_list = []
labels = []

# Loop through each domain
for domain in domains:
    # Loop through each category in the domain
    categories = os.listdir(os.path.join(base_directory, domain))
    for category in categories:
        category_directory = os.path.join(base_directory, domain, category)
        image_files = [f for f in os.listdir(category_directory) if f.endswith('.jpg')]

        
        for image_file in image_files:
            image_path = os.path.join(category_directory, image_file)
            example_image = cv2.imread(image_path)

            # Resize the image to a specific size
            example_image = cv2.resize(example_image, (128, 128))

            # Extract color histograms
            color_histogram = cv2.calcHist([example_image], [0, 1, 2], None, [8, 8, 8], [0, 256, 0, 256, 0, 256]).flatten()

            # Convert the image to grayscale
            gray_image = cv2.cvtColor(example_image, cv2.COLOR_BGR2GRAY)

            # Extract HOG features
            hog_features, hog_image = hog(gray_image, orientations=8, pixels_per_cell=(16, 16), cells_per_block=(1, 1), visualize=True, block_norm='L2-Hys')

            # Combine color histograms and HOG features
            combined_features = np.concatenate([color_histogram, hog_features])

            features_list.append(combined_features)

            # Record the label information (domain and category) for this image
            labels.append((domain, category))
```

```python
flattened_labels = ['_'.join(label) for label in labels]
```

```python
# Ensure unique labels
unique_labels = list(set(flattened_labels))

# Convert labels to numeric format
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(unique_labels)

# Map the unique labels back to the original labels
label_mapping = dict(zip(unique_labels, encoded_labels))
encoded_labels = [label_mapping[label] for label in flattened_labels]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features_list, encoded_labels, test_size=0.2, random_state=42)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Create an SVM classifier
svm_classifier = SVC(kernel='rbf', C=1.0, random_state=42)

svm_classifier.fit(X_train, y_train)

y_pred = svm_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
import joblib
from sklearn.ensemble import RandomForestClassifier

# Create a Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier
rf_classifier.fit(X_train, y_train)

# Predict labels for the test set
y_pred_rf = rf_classifier.predict(X_test)

# Evaluate the accuracy
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f'Random Forest Accuracy: {accuracy_rf * 100:.2f}%')
joblib.dump(rf_classifier, 'random_forest_model.pkl')
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Create a KNN classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)

# Train the classifier
knn_classifier.fit(X_train, y_train)

# Predict labels for the test set
y_pred_knn = knn_classifier.predict(X_test)

# Evaluate the accuracy
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f'KNN Accuracy: {accuracy_knn * 100:.2f}%')
```

```python
import matplotlib.pyplot as plt

# List of models and their accuracies
models = ['SVM', 'Random Forest', 'KNN']
accuracies = [accuracy * 100, accuracy_rf * 100, accuracy_knn * 100]

# Create a bar plot
plt.bar(models, accuracies, color=['blue', 'green', 'red'])
plt.ylabel('Accuracy (%)')
plt.title('Model Comparison')
plt.ylim([0, 100])  

# Display the accuracy values on top of the bars
for i, accuracy in enumerate(accuracies):
    plt.text(i, accuracy + 1, f'{accuracy:.2f}%', ha='center')

# Show the plot
plt.show()
```

```python

def retrieve_images_different_domains(base_directory, predicted_domain, predicted_category, num_images=3):
    # Get the list of all domains
    all_domains = os.listdir(base_directory)

    # Remove invalid domains and the predicted domain from the list
    invalid_domains = ['.ipynb_checkpoints']
    valid_domains = [domain for domain in all_domains if domain not in invalid_domains and domain != predicted_domain]

    # Shuffle the remaining domains
    np.random.shuffle(valid_domains)

    # Initialize a list to store the retrieved images along with domain and category information
    retrieved_images_info = []

    for domain in valid_domains[:num_images]:
        # Load the top image from the selected domain and category
        image_directory = os.path.join(base_directory, domain, predicted_category)
        image_files = [f for f in os.listdir(image_directory) if f.endswith('.jpg')]

        if image_files:
            # Select a random image from the list
            selected_image_file = np.random.choice(image_files)
            selected_image_path = os.path.join(image_directory, selected_image_file)

            # Read the image
            selected_image = cv2.imread(selected_image_path)

            # Append the image along with domain and category information to the list
            retrieved_images_info.append((selected_image, domain, predicted_category))

    return retrieved_images_info
```

```python
def display_images(query_image, retrieved_images_info):
    plt.figure(figsize=(15, 5))

    # Display the query image
    plt.subplot(1, 4, 1)
    plt.imshow(cv2.cvtColor(query_image, cv2.COLOR_BGR2RGB))
    plt.title('Query Image')
    plt.axis('off')

    # Display the retrieved images
    for i, (image, domain, category) in enumerate(retrieved_images_info):
        plt.subplot(1, 4, i + 2)
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.title(f'Retrieved Image {i + 1}\nDomain: {domain}\nCategory: {category}')
        plt.axis('off')

    plt.show()
```
